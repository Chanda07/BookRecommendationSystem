# Book Recommendation System

```python
#Import required libraries
import numpy as np
import pandas as pd

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics import confusion_matrix, plot_confusion_matrix
from sklearn import metrics
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Extraction of data from the csv files
df=pd.read_csv("books.csv")

#Read first ten records from the files
df.head(10) 
```

```python
df.info()
```

```python
#Reading the data from csv file and displaying 7 records
df1 = pd.read_csv('Ratings.csv')
df1.head(7)
```

```python
df1.Rating.value_counts()
```

```python
from sklearn.utils import resample
zero = df1[df1.Rating == 0]
one = df1[df1.Rating == 1]
```

```python
len(zero)
```

```python
df_final = resample(one, n_samples = len(zero))
df_Final = pd.concat([df_final,zero],axis = 0)
```

```python
df_Final.Rating.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X = df_Final.drop(columns = ['Rating'])
y = df_Final['Rating']
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size = 0.5, random_state = 42)
```

```python
x_train.shape
```

```python
cm_x = x_train[:1000]
cm_y = y_train[:1000]
```

```python
from sklearn.ensemble import RandomForestClassifier
RFC = RandomForestClassifier(random_state=43)
```

```python
model = RFC.fit(cm_x,cm_y)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
report = classification_report(y_test,y_pred)
```

```python
print(report)
```

```python
from sklearn.metrics import plot_confusion_matrix
import matplotlib.pyplot as plt
plot_confusion_matrix(model, x_test, y_test)
plt.show()
```

```python
#Select necessary columns from the csv file.
df=df[['title','authors','language_code','ratings_count']]
```

```python
#Data Preprocessing
#loc is utilized to pick rows by label or index.
#Select the value by the row and column label.
df.loc[(df['authors'].isnull()), 'authors'] = 'Unknown' 
df.loc[(df['language_code'].isnull()), 'language_code'] = 'Unknown'

#Converting data into the lower cases
df['language_code'] = df['language_code'].apply(lambda x: " ".join(x.lower() for x in x.split())) 
df['authors'] = df['authors'].apply(lambda x: " ".join(x.lower() for x in x.split()))

#Maintain all the information from the selected data column separate using commas in the "words".
df['words'] = "" 
df.loc[:,'words'] = df.loc[:, 'language_code'] +', '+ df.loc[:, 'authors']  

                                                                          
```

```python
#Read first ten records
df.head(10)
```

```python
#Creating a vector for the book datas

#The word count is specified and the sparse matrix got as a output.
vectorizer = TfidfVectorizer()
tfidf = vectorizer.fit_transform(df["words"]) 
```

```python
#Calculate similarity by cosine similarity
cosine_sim = cosine_similarity(tfidf, tfidf)
cosine_sim
```

```python
#Create function for recommendation
def recommendation(Title, cosine_sim):
    recommended = np.array([])
    index_ = df.loc[(df['title'] == Title)].index
    score = cosine_sim[index_]
    #Create a partition indirectly along the specified axis.
    indic = np.argpartition(score[0],[-10])[-8:] 
    #Carries out an indirect sort along the specified axis.
    indic = indic[np.argsort(score[0][indic])]
    return indic
                    
```

```python
#Recommending books similar to the book "The Hobbit"
indices = recommendation('The Hobbit', cosine_sim)
df[['title','authors','language_code','ratings_count']].loc[indices]
```

```python
#Recommending books similar to the book "The Fault in Our Stars"
indices = recommendation('The Fault in Our Stars', cosine_sim)
df[['title','authors','language_code','ratings_count']].loc[indices]
```
